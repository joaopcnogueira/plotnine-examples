```python
import numpy as np
import pandas as pd
from plotnine import *
from plotnine.data import diamonds
```

```python
diamonds.head()
```

# Basic Boxplot

```python
(
    ggplot(diamonds, aes(x="cut", y="price"))
    + geom_boxplot()
    + labs(x="Diamond cut", y="Price of diamond\n(USD)")
)
```

### Customising axis labels, title and changing axis ticks

```python
np.arange(0, 20001, 2500)
```

```python
(
    ggplot(diamonds, aes(x="cut", y="price"))

    # building the boxplot chart
    + geom_boxplot()

    # custom axis labels and title
    + labs(x="Diamond cut", 
           y="Price of diamond\n(USD)", 
           title="Price of diamonds by cut")
    
    # custom changing axis ticks
    + scale_y_continuous(breaks=np.arange(0, 20001, 2500),
                         limits=[0, 20000])
)
```

### Changing the colour of the boxes

```python
(
    ggplot(diamonds, aes(x="cut", y="price"))

    # building the boxplot chart
    + geom_boxplot(colour="rebeccapurple", fill="lightskyblue")

    # custom axis labels and title
    + labs(x="Diamond cut", 
           y="Price of diamond\n(USD)", 
           title="Price of diamonds by cut")
    
    # custom changing axis ticks
    + scale_y_continuous(breaks=np.arange(0, 20001, 2500),
                         limits=[0, 20000])
)
```

```python
(
    ggplot(diamonds, aes(x="cut", y="price"))

    # building the boxplot chart
    + geom_boxplot(colour="#1F3552", fill="#4271AE")

    # custom axis labels and title
    + labs(x="Diamond cut", 
           y="Price of diamond\n(USD)", 
           title="Price of diamonds by cut")
    
    # custom changing axis ticks
    + scale_y_continuous(breaks=np.arange(0, 20001, 2500),
                         limits=[0, 20000])
)
```

```python
(
    ggplot(diamonds, aes(x="cut", y="price"))

    # building the boxplot chart
    + geom_boxplot(colour="#1F3552", fill="#4271AE", alpha=0.7)

    # custom axis labels and title
    + labs(x="Diamond cut", 
           y="Price of diamond\n(USD)", 
           title="Price of diamonds by cut")
    
    # custom changing axis ticks
    + scale_y_continuous(breaks=np.arange(0, 20001, 2500),
                         limits=[0, 20000])
)
```

### Changing the appearance of the outliers

```python
(
    ggplot(diamonds, aes(x="cut", y="price"))

    # building the boxplot chart
    + geom_boxplot(colour="#1F3552", 
                   fill="#4271AE", 
                   alpha=0.7,
                   outlier_shape=".",
                   outlier_colour="steelblue")

    # custom axis labels and title
    + labs(x="Diamond cut", 
           y="Price of diamond\n(USD)", 
           title="Price of diamonds by cut")
    
    # custom changing axis ticks
    + scale_y_continuous(breaks=np.arange(0, 20001, 2500),
                         limits=[0, 20000])
)
```

### Changing the theme

```python
(
    ggplot(diamonds, aes(x="cut", y="price"))

    # building the boxplot chart
    + geom_boxplot(colour="#1F3552", 
                   fill="#4271AE", 
                   alpha=0.7,
                   outlier_shape=".",
                   outlier_colour="steelblue")

    # custom axis labels and title
    + labs(x="Diamond cut", 
           y="Price of diamond\n(USD)", 
           title="Price of diamonds by cut")
    
    # custom changing axis ticks
    + scale_y_continuous(breaks=np.arange(0, 20001, 2500),
                         limits=[0, 20000])
    
    # custom theme
    + theme_bw()
    # + theme_538()
)
```

# Grouping by Another Variable

```python
diamonds["carat_range"] = pd.qcut(diamonds["carat"], 2, labels=["Lower carat", "Higher carat"])
```

```python
(
    ggplot(diamonds, aes(x="cut", y="price"))

    # building the boxplot chart
    + geom_boxplot(colour="#1F3552", 
                   fill="#4271AE", 
                   alpha=0.7,
                   outlier_shape=".",
                   outlier_colour="steelblue")

    # custom axis labels and title
    + labs(x="Diamond cut", 
           y="Price of diamond\n(USD)", 
           title="Price of diamonds by cut")
    
    # custom changing axis ticks
    + scale_y_continuous(breaks=np.arange(0, 20001, 2500),
                         limits=[0, 20000])
    
    # custom theme
    + theme_bw()
    
    + facet_grid(". ~ carat_range")
)
```

### Alterando o tamanho da figura

```python
import plotnine

# tamanho padrão da figura
plotnine.options.figure_size
```

[Source code for plotnine.options](https://plotnine.readthedocs.io/en/stable/_modules/plotnine/options.html)

```python
# aumentando o tamanho padrão da figura
plotnine.options.figure_size = (10, 4.8)

(
    ggplot(diamonds, aes(x="cut", y="price"))
    + geom_boxplot(colour="#1F3552", 
                   fill="#4271AE", 
                   alpha=0.7,
                   outlier_shape=".",
                   outlier_colour="steelblue")
    + labs(x="Diamond cut", 
           y="Price of diamond\n(USD)", 
           title="Price of diamonds by cut")
    + scale_y_continuous(breaks=np.arange(0, 20001, 2500),
                         limits=[0, 20000])
    + theme_bw()
    + facet_grid(". ~ carat_range")
)
```
